## INTRODUCTION

I have selected a dataset for analysis, focusing on an individual shooting incident that occurred in New York City between 2006 and the end of the prior calendar year. This dataset contains details about the incident's date, time, location, and specifics, along with demographic information about the person involved.

The research question to answer with this data is : What patterns and trends can be observed in shooting incidents in New York City from 2006 to the conclusion of the previous calendar year, and what are the demographic elements linked to these incidents?

By scrutinizing this data, the aim is to unveil any possible recurring pattern or trend in the incidence and geographical distribution of shooting incidents. At the same time, the goal is to explore potential associations between these occurrences and demographic factors such as age, gender, and race. The insights obtained from this analysis can offer valuable direction for shaping policy and implementing strategies to tackle and diminish gun violence in New York City.

The dataset containing information about shooting incidents in New York City from 2006 to 2022 was acquired from the NYC Open Data website (https://data.cityofnewyork.us/Public-Safety/NYPD-Shooting-Incident-Data-Historic-/833y-fsy8). This dataset furnishes details about shooting occurrences within the city, encompassing data regarding the date, time, location, the count of victims, and the type of weapon employed.

## DATA SUMMARY

The dataset comprises 25,596 individual instances and encompasses 19 attributes. Each instance represents a distinct shooting incident, and these attributes provide details about the incident, including its date and time, location, and the demographics of both the perpetrator and victim. Additionally, it includes geographical coordinates for each incident.

Here are the attribute data types:

INCIDENT_KEY: An integer, serving as a unique, randomly generated ID for each arrest.

OCCUR_DATE: A string/object representing the precise date of the shooting incident.

OCCUR_TIME: A string/object indicating the exact time of the shooting incident.

BORO: A string/object specifying the borough where the shooting incident occurred.

PRECINCT: An integer denoting the precinct where the shooting incident took place.

JURISDICTION_CODE: A float representing the jurisdiction in which the shooting incident occurred. Jurisdiction codes 0 (Patrol), 1 (Transit), and 2 (Housing) pertain to the NYPD, while codes 3 and above refer to non-NYPD jurisdictions.

LOCATION_DESC: A string/object describing the location of the shooting incident.

STATISTICAL_MURDER_FLAG: A boolean value indicating whether the shooting resulted in the victim's death, which would be classified as a murder.

PERP_AGE_GROUP: A string/object categorizing the age group of the perpetrator.

PERP_SEX: A string/object describing the gender of the perpetrator.

PERP_RACE: A string/object indicating the race of the perpetrator.

VIC_AGE_GROUP: A string/object categorizing the age group of the victim.

VIC_SEX: A string/object specifying the gender of the victim.

VIC_RACE: A string/object indicating the race of the victim.

X_COORD_CD: A float representing the midblock X-coordinate in the New York State Plane Coordinate System, Long Island Zone, NAD 83, measured in feet (FIPS 3104).

Y_COORD_CD: A float denoting the midblock Y-coordinate in the New York State Plane Coordinate System, Long Island Zone, NAD 83, measured in feet (FIPS 3104).

Latitude: A float representing the latitude coordinate in the Global Coordinate System, WGS 1984, in decimal degrees (EPSG 4326).

Longitude: A float indicating the longitude coordinate in the Global Coordinate System, WGS 1984, in decimal degrees (EPSG 4326).

Lon_Lat: A string/object representing longitude and latitude coordinates for mapping purposes.

```python
# To get the summary of OCCUR_DATE
print(shooting_data['OCCUR_DATE'].describe())
```

This summary statistics pertains to the 'OCCUR_DATE' attribute in the NYPD dataset. There are a total of 25,596 entries within this column, and these entries comprise 5,409 unique values. The most frequently occurring date is July 5th, 2020, appearing 47 times. The dataset's earliest recorded date is January 1st, 2006, while the latest date is December 31st, 2021. The FutureWarning message arises because the 'OCCUR_DATE' column contains datetime data, which is currently treated as categorical data in the describe() function. To prevent this warning and adopt the upcoming behavior, the warning advises using datetime_is_numeric=True.

To conduct a visual analysis of the OCCUR_DATE attribute within the nypd_data DataFrame, employing Matplotlib and Seaborn, we can initiate by extracting the year, month, and day components from the OCCUR_DATE column via the pd.to_datetime method. Subsequently, these extracted details can be employed to generate visual representations illustrating the temporal distribution of incidents.

Here is the Python code that accomplishes the extraction of year, month, and day information from the OCCUR_DATE column and constructs the plots using Matplotlib and Seaborn:

```python
# Converting OCCUR_DATE column to datetime format
shooting_data['OCCUR_DATE'] = pd.to_datetime(shooting_data['OCCUR_DATE'])

# Extracting year, month, and day information from OCCUR_DATE
shooting_data['YEAR'] = shooting_data['OCCUR_DATE'].dt.year
shooting_data['MONTH'] = shooting_data['OCCUR_DATE'].dt.month
shooting_data['DAY'] = shooting_data['OCCUR_DATE'].dt.day
```

```python
# Visualizing the yearly distribution of crimes using Matplotlib

# Setting the plot size
plt.figure(figsize=(6, 5))

# Creating a bar plot showing the number of crimes per year, sorted by year
shooting_data['YEAR'].value_counts().sort_index().plot(kind='bar',color='green')

# Setting the title of the plot
plt.title('Crime Distribution Over the Years')

# Labeling the x-axis
plt.xlabel('Year')

# Labeling the y-axis
plt.ylabel('Number of Crimes')

# Displaying the plot
plt.show()
```

The data indicates that crime rates were highest in the year 2006, followed by peaks in 2008 and 2011. Subsequently, there was a decrease in crime rates, with the lowest recorded in 2019 at approximately 1,000 incidents per year. However, in the last couple of years, 2020 and 2021, crime rates appear to have surged again, reaching approximately 2,000 incidents per year. In the intervening years between 2006 and 2021, crime rates generally ranged from 1,250 to 1,850 incidents annually.

```python
# Visualizing the distribution of incidents by month using Seaborn

# Adjusting the plot size
plt.figure(figsize=(7, 5))

# Creating a count plot with Seaborn and specifying the color
sns.countplot(x='MONTH', data=shooting_data, color='orange')

# Adding a title to the plot
plt.title('Incident Distribution by Month')

# Labeling the x-axis
plt.xlabel('Month')

# Labeling the y-axis
plt.ylabel('Number of Incidents')

# Displaying the plot
plt.show()
```

This code produces a Seaborn count plot to visualize the distribution of crimes over different months. The plot's dimensions are set with a width of 7 units and a height of 5 units. It creates a count plot using Seaborn, illustrating the count of crimes for each month on the x-axis. The plot is labeled "Number of Crimes by Month," and the x and y-axes are appropriately labeled. Finally, the plot is displayed for viewing.

```python
# Visualizing the frequency of incidents by day using Seaborn

# Adjusting the plot size
plt.figure(figsize=(10, 5))

# Creating a count plot with Seaborn
sns.countplot(x='DAY', data=shooting_data,color='pink')

# Adding a title to the plot
plt.title('Incident Count by Day of Month')

# Labeling the x-axis
plt.xlabel('Day of Month')

# Labeling the y-axis
plt.ylabel('Number of Incidents')

# Displaying the plot
plt.show()

```

The most significant number of incidents typically occurs on the 5th day of the month, with approximately 980 incidents per day, while the fewest incidents are recorded on the last day of the month, totaling around 600 incidents daily. To visualize how the frequency of shooting incidents has changed over time, we can create a line plot using the occur_date column. First, we need to convert the occur_date column to a datetime format so that we can easily group the data by year and month.
